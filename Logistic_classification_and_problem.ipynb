```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.array([(0,0),(0,1),(1,0),(1,1)])
y = np.array([[0,0,0,1]]).reshape(-1,1)
```

```python
plt.scatter(x[0:3, 0], x[0:3, 1], c='red', edgecolors='none', s=30, label='0')
plt.scatter(x[3, 0], x[3, 1], c='blue', edgecolors='none', s=30, label='1')
plt.legend(loc=1)
plt.xlabel('X1')
plt.ylabel('X2')
```

```python
x = np.hstack((np.ones((4,1)),x))
w = np.array([0,1,1]).reshape(-1,1)
```

```python
numIter = 1000
cost = np.zeros((numIter, 1))
learningrate = 1
for i in range(1,numIter):
    y_predict = sigmoid(np.dot(x,w))
    cost[i] = -np.sum(np.multiply(y, np.log(y_predict))+\
                      np.multiply(1-y, np.log(1-y_predict)))
    #gradient descent
    w = w-learningrate*np.dot(x.T, y_predict-y)
    print('Epoch {} Cost: {:.6f}'.format(
            i, cost[i,0]
        ))
```

```python
t = 0.5
x = np.array([(0,0),(0,1),(1,0),(1,1)])
y = np.array([[0,0,0,1]]).reshape(-1,1)
plt.scatter(x[0:3, 0], x[0:3, 1], c='red', edgecolors='none', s=30, label='0')
plt.scatter(x[3, 0], x[3, 1], c='blue', edgecolors='none', s=30, label='1')
plt.legend(loc=1)
plt.xlabel('X1')
plt.ylabel('X2')
plt.plot((0, 1),(-(w[0]+0*w[1]+ np.log(1/t-1))/w[2], -(w[0] + 1*w[1]+ np.log(1/t-1))/w[2]), 'g')
plt.show()
```
